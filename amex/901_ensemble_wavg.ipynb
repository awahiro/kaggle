```python
import pandas as pd
import os
import pprint


pprint.pprint(os.listdir("./output/submission"))

```

```python
# 加重平均で算出 weighted average
# 加重平均は、
# 値, 重みが、
# A, a
# B, b
# C, c
# とした場合、
# 加重平均 = (A*a + B*b + C*c) / (a+b+c)
# 

# 対象ファイル
file_names = [
#     'sub_cross_validation.csv',
#     'sub_benchmark_random-rorest-regressor.csv',
#     'sub_cross_validation_with_last_selectkbest_custom-metric_param-dart.csv',
#     'sub_cross_validation_with_last_selectkbest_param-dart.csv',
#     'sub_benchmark_random-rorest-classifier.csv',
#     'sub_cross_validation_random-forest_with_last_selectkbest.csv',
#     'sub_benchmark_plus_xboost.csv',
#     'sub_cross_validation_with_selectkbest.csv',
    ('sub_cross_validation_random-forest_with_last_selectkbest_param-tune.csv', 0.770),
    ('sub_cross_validation_with_last_selectkbest_custom-metric.csv', 0.783),
#     'sub_benchmark_plus_light-gbm.csv',
    ('sub_cross_validation_nn_with_last_selectkbest.csv', 0.787),
    ('sub_cross_validation_with_last_selectkbest_param-tune.csv', 0.789),
#     'sub_benchmark_plus_nn_2.csv',
    ('sub_cross_validation_xboost_with_last_selectkbest_param-tune-dart.csv', 0.783),
#     'sub_benchmark_random-rorest.csv',
#     ('sub_cross_validation_xboost_with_last_selectkbest.csv', 0.785),
#     'sub_cross_validation_plus.csv',
#     ('sub_cross_validation_with_last_selectkbest.csv', 0.786),
#     'sub_cross_validation_nn_with_last_selectkbest_param-tune.csv',
#     'sub_plus_custom-metric.csv',
#     'sub_benchmark_plus_nn.csv',
#     'sub_benchmark_plus_random-rorest-regressor.csv',
#     'sub_benchmark_light-gbm.csv'
]

# 平均を計算
sub = pd.read_csv('input/sample_submission.csv')
file_dfs = []
weight_sum = 0
for (file_name, weight) in file_names:
    file_df = pd.read_csv(f'./output/submission/{file_name}')
    print('file_name',file_name,  file_df.shape)
    sub['prediction'] = sub['prediction'] + file_df['prediction'] * weight
    print('weight:', weight)
    print('file_df:', file_df)
    weight_sum += weight

print('weight_sum:', weight_sum)

sub['prediction'] =  (sub['prediction'] /weight_sum).round(2)

sub
```

```python
# ファイル出力

# prefix
prefix_parts = []
for (file_name, weight) in file_names:
    tmp = file_name.replace('sub_', '').replace('.csv', '')
    prefix_parts.append(tmp)
prefix_parts.sort()
prefix = '__'.join(prefix_parts)
prefix = prefix[:200]

import datetime
now = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
output_file = f'./output/ensemble/esb_wavg_{now}_{prefix}.csv'

sub.to_csv(output_file, index=False)
print(output_file)
sub
```
