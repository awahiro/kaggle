### benchmark：xboost

```python
import gc
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import pandas_profiling as pp


warnings.filterwarnings('ignore')
%matplotlib inline


print(os.curdir)
print(os.listdir("input"))
```

```python
%%time
h5File = "temp_data/h5/prepared_data_with_last.h5";
key_train = 'prepared_01/train';
key_valid = 'prepared_01/valid';
key_test = 'prepared_01/test';

train = pd.read_hdf(h5File, key_train); gc.collect();
valid = pd.read_hdf(h5File, key_valid); gc.collect();
test = pd.read_hdf(h5File, key_test); gc.collect();
```

```python
X_train = train
y_train = valid
X_test = test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.3, random_state=0, stratify=y_train)
print(X_train.head())
print(X_valid.head())
print(y_train.head())
print(y_valid.head())
```

```python
%%time

from xgboost import XGBClassifier
model = XGBClassifier(early_stopping_rounds=10)
eval_set = [(X_valid, y_valid)]
model.fit(X_train, y_train, eval_set=eval_set, verbose=True)
```

```python
%%time
y_pred = model.predict_proba(X_test)
```

```python
y_pred.shape
```

```python
sub = pd.read_csv('input/sample_submission.csv')
sub
```

```python
# create submission
X_tiny = X_test[['count']]
X_tiny = pd.merge(X_tiny, sub, left_index=True, right_on='customer_ID')
X_tiny['prediction'] = 1 - y_pred.round(2) # xboostは反転が必要そう
X_tiny = X_tiny.drop(columns='count')
sub = X_tiny
sub
```

```python
sub.to_csv('output/submission/sub_benchmark_plus_xboost.csv', index=False)
```
