```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import pandas_profiling as pp
import joblib
import subprocess

warnings.filterwarnings('ignore')
%matplotlib inline

print(os.curdir)
print(os.listdir("input"))
```

```python
def get_train_data(num=1):
    max = 7
    i = 0
    file_name = f'temp_data/agg/train_data_agg_0000000{i:05}.csv'
    print(file_name)
    df = pd.read_csv(file_name, index_col=0)
    i = 1
    for i in range(1, num):
        if i > max: 
            break
        file_name = f'temp_data/agg/train_data_agg_0000000{i:05}.csv'
        print(file_name)
        tmp = pd.read_csv(file_name, index_col=0)
        df = pd.concat([df, tmp])
        i = i + 1

        return df

def get_test_data(num=1):
    max = 13
    i = 0
    file_name = f'temp_data/agg/test_data_agg_0000000{i:05}.csv'
    print(file_name)
    df = pd.read_csv(file_name, index_col=0)
    i = 1
    for i in range(1, num):
        if i > max: 
            break
        file_name = f'temp_data/agg/test_data_agg_0000000{i:05}.csv'
        print(file_name)
        tmp = pd.read_csv(file_name, index_col=0)
        df = pd.concat([df, tmp])
        i = i + 1

    return df

def get_train_valid_data():
    df = pd.read_csv('input/train_labels.csv', index_col=0)
    return df
```
