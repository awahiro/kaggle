```python
import numpy as np 
import pandas as pd

####################################################################
# test データ取得
####################################################################

# データファイル分割のprefixリストを取得する
# split_digit : 分割するprefixの桁数
def get_test_prefixes(split_digit): 
    test_prefixes = pd.read_csv('temp_data/test_data/test_uniq_prefix_2.csv', header=None)
    prefixes = test_prefixes[0].str[:split_digit].unique()
    prefixes = list(prefixes)
    return prefixes


# test の prefixを指定してdfを取得する
def get_test_data_df_by_prefixes(prefixes):
    print(f'prefixes={prefixes}')
    # 事前に 作っておく、head -n 1 ./input/test_data.csv > ./temp_data/header_of_test_data.csv
    df = pd.read_csv(f'./temp_data/header_of_test_data.csv', index_col=0)

    for prefix in prefixes:
        tmp = pd.read_csv(f'temp_data/test_data/p2/test_{prefix}.csv', header=None, index_col=0)
        tmp.columns = df.columns
        tmp.index.name = df.index.name
        df = pd.concat([df, tmp])
    return df

# train の prefixを指定してdfを取得する
def get_test_data_df(pick_num=0, pos=0, random=False):
    
    prefixes = get_test_prefixes(2)
    
    if random:
        import random
        random.seed(1)
        prefixes = random.sample(prefixes, len(prefixes))
    
    if pick_num > 0:
        prefixes = prefixes[pos: pos + pick_num]
    
    return get_test_data_df_by_prefixes(prefixes)


####################################################################
# train データ取得
####################################################################

# データファイル分割のprefixリストを取得する
# split_digit : 分割するprefixの桁数
def get_train_prefixes(split_digit): 
    train_labels = pd.read_csv('input/train_labels.csv')
    prefixes = train_labels['customer_ID'].str[:split_digit].unique()
    prefixes = list(prefixes)
    return prefixes

# train の prefixを指定してdfを取得する
def get_train_data_df_by_prefixes(prefixes):
    print(f'prefixes={prefixes}')
    # 事前に 作っておく、head -n 1 ./input/train_data.csv > ./temp_data/header_of_train_data.csv
    df = pd.read_csv(f'./temp_data/header_of_train_data.csv', index_col=0)

    for prefix in prefixes:
        tmp = pd.read_csv(f'temp_data/train_data/p2/train_{prefix}.csv', header=None, index_col=0)
        tmp.columns = df.columns
        tmp.index.name = df.index.name
        df = pd.concat([df, tmp])
    return df

# train の prefixを指定してdfを取得する
def get_train_data_df(pick_num=0, pos=0, random=False):
    
    prefixes = get_train_prefixes(2)
    
    if random:
        import random
        random.seed(1)
        prefixes = random.sample(prefixes, len(prefixes))
    
    if pick_num > 0:
        prefixes = prefixes[pos: pos + pick_num]
    
    return get_train_data_df_by_prefixes(prefixes)

```
