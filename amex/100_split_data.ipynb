```python
def __mkdir(path):
    if not os.path.exists(path):
        os.mkdir(path)
    
__mkdir('temp_data')
__mkdir('temp_data/train_data')
__mkdir('temp_data/train_data/p1')
__mkdir('temp_data/train_data/p2')
__mkdir('temp_data/test_data')
__mkdir('temp_data/test_data/p1')
__mkdir('temp_data/test_data/p2')
```

```python
###############################
# train_data
###############################
```

```python
# prefixでデータファイルを分割するためのprefixリストをデータファイルから取得する
# split_digit : 分割するprefixの桁数
def get_train_prefixes(split_digit): 
    train_labels = pd.read_csv('input/train_labels.csv')
    prefixes = train_labels['customer_ID'].str[:split_digit].unique()
    prefixes = list(prefixes)
    return prefixes
```

```python
%%time
# train_dataの分割：親ファイル
# prefix=1桁で親ファイルを作成する
# 並列処理
# 各prefixごとのデータファイルを作成する
cpu = 4

def split_data(prefix):
    cmd = f'time grep -e  "^{prefix}[0-f]" ./input/train_data.csv > ./temp_data/train_data/p1/train_{prefix}.csv'
    proc = subprocess.run(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    return f'--- {prefix} ------------\n{proc.stdout}\n{proc.stderr}'

split_digit = 1
prefixes = get_train_prefixes(split_digit)
prefixes = prefixes[:]
print(f'prefixes={prefixes}')
results = joblib.Parallel(n_jobs=cpu)(joblib.delayed(split_data)(prefix) for prefix in prefixes)

for result in results:
    print(result)
```

```python
%%time
# train_dataの分割：子ファイル
# prefix=2桁で子ファイルを作成する
# 並列処理
# 各prefixごとのデータファイルを作成する
cpu = 4

def split_data(prefix):
    digit_1 = prefix[0]
    cmd = f'time grep -e  "^{prefix}[0-f]" ./temp_data/train_data/p1/train_{digit_1}.csv > ./temp_data/train_data/p2/train_{prefix}.csv'
    proc = subprocess.run(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    return f'--- {digit_1} - {prefix} ------------\n{proc.stdout}\n{proc.stderr}'
#     return f'--- {digit_1} - {prefix} -----\n{cmd}'


split_digit = 2
prefixes = get_train_prefixes(split_digit)
prefixes = prefixes[:]
print(f'prefixes={prefixes}')
results = joblib.Parallel(n_jobs=cpu)(joblib.delayed(split_data)(prefix) for prefix in prefixes)

for result in results:
    print(result)
```
